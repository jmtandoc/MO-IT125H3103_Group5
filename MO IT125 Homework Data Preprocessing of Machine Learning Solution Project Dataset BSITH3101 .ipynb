```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Customer Demographics Data Cleaning

```python
#Loading customer_demogrphics contaminated data

cust_demo = pd.read_csv(r"C:\Users\jhoan\Downloads\customer_demographics_contaminated.csv")

#Showing only first 10 entries
cust_demo.head(10)
```

```python
#Removing Duplicates

cust_demo.drop_duplicates(inplace=True)
cust_demo
```

```python
#Showing info summary

cust_demo.info()
```

```python
#Cleaning Age Column - Changing column to  numeric/int value. Getting Median age and replacing null values to median value.

cust_demo['Age'] = pd.to_numeric(cust_demo['Age'], errors='coerce')
median_age = cust_demo['Age'].median()
cust_demo['Age'].fillna(median_age, inplace=True)
cust_demo['Age'] = cust_demo['Age'].astype(int)
cust_demo
```

```python
#Replacing negative values with Median Age

cust_demo['Age']= cust_demo['Age'].replace(-1, median_age)
cust_demo
```

```python
#Replacing missing values with Unknown

cust_demo['IncomeLevel'] = cust_demo['IncomeLevel'].fillna('Unknown')
cust_demo['IncomeLevel'].value_counts()
```

```python
# Replacing Outliers with 95th Percentile
percentile_95 = cust_demo['Age'].quantile(0.95)
cust_demo['Age'] = np.where(cust_demo['Age']>percentile_95, percentile_95, cust_demo['Age'])
cust_demo
```

```python
#Sign-up data to datetime datatype

cust_demo['SignupDate'] = pd.to_datetime(cust_demo['SignupDate'], format='mixed')
cust_demo
```

```python
# Cleaned data to CSV

cust_demo.to_csv(r"C:\Users\jhoan\Downloads\Jupyter Output Files\cleaned_cust_demo.csv")
```

# Customer_Transactions Data Cleaning

```python
#Loading Customer Transactions Data and Just showing first 10

cust_trans = pd.read_csv(r"C:\Users\jhoan\Downloads\customer_transactions_contaminated.csv")
cust_trans.head(10)
```

```python
cust_trans.drop_duplicates(inplace=True)
cust_trans
```

```python
#Show summary of data
cust_trans.info()
```

```python
#Replace Free to 0
cust_trans['Amount'] = cust_trans['Amount'].replace('Free',0)
cust_trans['Amount'] = cust_trans['Amount'].replace('',0)
cust_trans['Amount'] = cust_trans['Amount'].astype(float)
cust_trans
```

```python
#Replace -100 with 5th Percentile
percentile_5 = cust_trans['Amount'].quantile(0.05)
cust_trans['Amount'] = cust_trans['Amount'].replace(-100,percentile_5)
cust_trans
```

```python
#Replacing missing values with Median
cust_trans['Amount'] =cust_trans['Amount'].fillna(cust_trans['Amount'].median())
amount_median = cust_trans['Amount'].median()
cust_trans['Amount'] =cust_trans['Amount'].fillna(amount_median)
cust_trans
```

```python
#Replacing null values with Unknown
cust_trans['ProductCategory'] = cust_trans['ProductCategory'].fillna('Unknown')
cust_trans
```

```python
#Convert Datetime to datatype
cust_trans['TransactionDate'] = pd.to_datetime(cust_trans['TransactionDate'], format='mixed')
cust_trans
```

```python
#Cleaned Data to CSV

cust_trans.to_csv(r"C:\Users\jhoan\Downloads\Jupyter Output Files\cleaned_cust_trans.csv")
```

# Social Media Interactions Data Cleaning

```python
#Loading Social Media Interactions Dataset

soc_med = pd.read_csv(r"C:\Users\jhoan\Downloads\social_media_interactions_contaminated.csv")
soc_med.head(10)
```

```python
#Removing Duplicates

soc_med.drop_duplicates(inplace=True)
soc_med
```

```python
#Showing data summary

soc_med.info()
```

```python
#Cleaned Data to CSV

soc_med.to_csv(r"C:\Users\jhoan\Downloads\Jupyter Output Files\cleaned_soc_med.csv")
```
